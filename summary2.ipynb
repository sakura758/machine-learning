np.hstack([X, X[:,1:]**2])

# 2. 按位取反运算符

在Python中，~ 是按位取反运算符（Bitwise NOT Operator）。它会对一个整数的二进制表示进行按位取反操作，即将每一位的 0 变为 1，1 变为 0。

例如，假设有一个二进制数 b，它的形式是 1010，那么 ~b 的结果将是 0101。

如果应用在整数上，~ 会对整数的所有位进行取反，包括符号位。这意味着它会改变整数的符号并且增加一个单位。

```python
a = 5  # 二进制表示是 0101
b = ~a  # 按位取反后，得到 1010
```

```python
print(a)
print(b)
```

# 3. ravel()

ravel 是一个NumPy库中的函数，用于将多维数组（如矩阵）转换成一维数组。这个函数返回一个扁平化的数组，其中包含了原始数组中的所有元素。

ravel 是一个NumPy库中的函数，用于将多维数组（如矩阵）转换成一维数组。这个函数返回一个扁平化的数组，其中包含了原始数组中的所有元素。

```python
import numpy as np

# 创建一个二维数组
arr = np.array([[1, 2, 3],
                [4, 5, 6]])

# 使用 ravel 将其扁平化
flattened_arr = np.ravel(arr)

print(flattened_arr)
```

与 ravel 类似的函数还有 flatten，它也可以将多维数组转换成一维数组。但是有一个重要的区别：ravel 返回的是原数组的视图（view），而 flatten 返回的是原数组的拷贝（copy）。

flattened_arr = arr.flatten()

因此，如果你修改 flattened_arr 中的元素，它可能会影响到原始数组 arr（对于 ravel 而言），但不会影响到原始数组（对于 flatten 而言）。总的来说，如果你只需要读取扁平化后的数组，并且对性能要求比较高，可以使用 ravel。如果你需要对扁平化后的数组进行修改，并且希望不影响原始数组，那么可以使用 flatten。

# 4. np.linspace

np.linspace 是 NumPy 库中的一个函数，用于在指定的间隔内生成等间隔的数列。

```python
import numpy as np

# 生成一个包含5个元素的等间隔数列，起始值是1，终止值是10
# 这个数列包括了1和10
# 默认情况下，生成的数列是闭区间的（包括起始值和终止值）
arr = np.linspace(1, 10, 5)

print(arr)
```

np.linspace(start, stop, num) 接受以下参数：

start: 数列的起始值。       
stop: 数列的终止值。       
num: 要生成的等间隔样本数量。      
np.linspace 默认生成的数列是闭区间的（包括起始值和终止值）。如果你希望生成半开区间（包括起始值但不包括终止值），你可以使用 endpoint=False 参数。

# 6. meshgrid

meshgrid 是 NumPy 中的一个函数，它用于生成一个网格坐标矩阵，通常用于在二维平面中生成坐标点。

```python
import numpy as np

x = np.array([1, 2, 3])
y = np.array([4, 5, 6, 7])

X, Y = np.meshgrid(x, y)
print(X)
print(Y)
```

在上述代码中，x 和 y 是两个一维数组。meshgrid 会生成两个矩阵 X 和 Y，其中 X 的行数是 y 中元素的个数，列数是 x 中元素的个数，Y 的行数是 y 中元素的个数，列数是 x 中元素的个数。

X 中的每一行都是 x 中的元素的复制；    
Y 中的每一列都是 y 中的元素的复制。    
这样，X 和 Y 就构成了一个网格，其中包含了所有可能的坐标点。

# 7. plot_surface()

plot_surface 是 Matplotlib 库中用于绘制三维曲面图的函数。

```python
import matplotlib.pyplot as plt
import numpy as np

# 创建一个网格
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
X, Y = np.meshgrid(x, y)

# 生成一个二元函数 z = f(x, y) 的值
Z = np.sin(np.sqrt(X**2 + Y**2))

# 绘制三维曲面图
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
surf = ax.plot_surface(X, Y, Z, cmap='viridis')

# 添加颜色条
fig.colorbar(surf)

# 显示图形
plt.show()
```

在这个例子中，我们首先创建了一个网格，然后通过二元函数 z = f(x, y) 生成了一个与 x 和 y 相关的 z 值。接着，使用 ax.plot_surface 绘制了三维曲面图。

plot_surface 接受的参数包括：

X 和 Y：分别是 x 和 y 坐标的网格矩阵。
Z：是与 x 和 y 对应的 z 值的网格矩阵。
cmap：指定颜色映射，决定了曲面的颜色。
此外，你可以通过修改 cmap 参数来改变曲面的颜色映射。

需要注意的是，这里用到了 projection='3d' 来指定了三维坐标系。这是因为我们要绘制的是一个三维曲面图。

ax.plot_surface(X, Y, Z, rstride=1, cstride=1, cmap=None, facecolors=None, norm=None, vmin=None, vmax=None, shade=False, **kwargs)


以下是这些参数的说明：

X, Y, Z: 分别是 x、y、z 坐标的网格矩阵。它们定义了曲面的形状。

rstride: 行步进，控制绘制曲面时的行步进大小。默认值为 1。

cstride: 列步进，控制绘制曲面时的列步进大小。默认值为 1。

cmap: 颜色映射，用于定义曲面的颜色。例如，'viridis'、'plasma' 等。默认为 None，表示使用默认的颜色映射。

facecolors: 曲面的颜色，可以用于覆盖 cmap。通常是一个与 Z 同样形状的数组。

norm: 用于映射颜色到数据值的标准化对象。

vmin, vmax: 用于设置颜色映射的最小值和最大值。

shade: 是否对曲面进行阴影处理。默认为 False。

**kwargs: 其他可选的参数，用于定制曲面的外观，例如颜色、透明度等。

plot_surface 用于绘制三维曲面图，其中 X, Y, Z 分别是网格矩阵，它们定义了曲面的形状和位置。通过调整 rstride 和 cstride 可以控制曲面的绘制精度。 cmap 用于定义曲面的颜色映射，而 facecolors 可以用来覆盖颜色映射。其他参数可以用于进一步定制曲面的外观。

# 9.decision_function()


在 Python 中，该decision_function()方法通常与支持向量机 (SVM) 分类器相关。此方法用于获取数据集中每个样本的决策值或分数。这些分数表示样本到超平面的有符号距离，可用于评估分类的置信度。

决策函数的确切值和解释取决于所使用的特定 SVM 变体、内核和参数。

```python
from sklearn import svm
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split

# Load the iris dataset for demonstration
iris = load_iris()
X_train, X_test, y_train, y_test = train_test_split(iris.data, iris.target, test_size=0.2, random_state=42)

# Create an SVM classifier
svm_classifier = svm.SVC(kernel='linear', probability=True)  # 'linear' kernel for simplicity

# Train the classifier
svm_classifier.fit(X_train, y_train)

# Use decision_function on the test set
decision_values = svm_classifier.decision_function(X_test)

# Print the decision values for the first few samples
print(decision_values[:5])
```

# 10. contourf()

contourf是 Python 中的一个函数，通常用于在数据可视化中创建填充等高线图。它主要与 Matplotlib 等库一起使用，用于创建数据的 2D 等高线图。等高线图的“填充”方面意味着它们将等高线表示为带有颜色的填充区域，从而更容易可视化数据分布。

```python
import numpy as np
import matplotlib.pyplot as plt

# Create sample data for contour plotting
x = np.linspace(-2, 2, 100)
y = np.linspace(-2, 2, 100)
X, Y = np.meshgrid(x, y)
Z = X**2 + Y**2  # Example function to contour

# Create a filled contour plot
plt.contourf(X, Y, Z, levels=20, cmap='viridis')
plt.colorbar()  # Add a colorbar to indicate values

# Add labels and a title
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Filled Contour Plot Example')

# Show the plot
plt.show()
```

在这个例子中：

我们导入必要的库，包括 NumPy 和 Matplotlib。

我们创建样本数据（X、Y和Z）来定义等高线图。X和Y是表示 x 和 y 坐标的二维数组，Z是表示每个点的数据值的二维数组。

我们用来plt.contourf创建填充等高线图。X、Y、 和Z作为参数传递。levels确定轮廓级别的数量，并cmap指定填充区域的颜色图。

plt.colorbar()用于添加颜色条，为图中不同颜色表示的值提供视觉参考。

plt.xlabel()我们使用、plt.ylabel()和向轴添加标签并为绘图添加标题plt.title()。

最后，我们使用 显示绘图plt.show()。

该contourf函数用途广泛，通过指定适当的数据和设置，可用于可视化各种数据，例如热图、地形数据等。
