# 聚类算法的对比与评估

## 1. 用真实值评估聚类

有一些指标可用于评估聚类算法相对于真实聚类的结果，其中最重要的是**调整rand 指数**(adjusted randindex，ARI)和**归一化互信息** (normalized mutualinformation，NMI)，二者都给出了定量的度量，其最佳值为1，0表示不相关的聚类(虽然ARI可以取负值)。

使用ARI来比较k 均值、凝聚聚类和DBSCAN 算法,添加了将点随机分配到两个簇中的图像:

```python
# 利用无监督的轮廓分数在twomoons数据集上比较随机分配k均值、凝聚聚类和DBSCAN(更符合直觉的DBSCAN的轮廓分数低于k均值找到的分配)
from sklearn.metrics.cluster import adjusted_rand_score
from sklearn.datasets import make_moons
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans, AgglomerativeClustering, DBSCAN
from sklearn.metrics import silhouette_score
import matplotlib.pyplot as plt
import numpy as np
import mglearn

X, y = make_moons(n_samples=200, noise=0.05, random_state=0)
# 将数据缩放成平均值为0，方差为1
scaler = StandardScaler()
scaler.fit(X)
X_scaled = scaler.transform(X)

''' subplot_kw={'xticks':(), 'yticks':()} 是用于配置 Matplotlib 中子图（subplot）的关键字参数（keyword arguments）。
当这些参数被设置为空时，Matplotlib 会使用默认的行为来确定刻度的位置。这意味着 Matplotlib 将会自动选择合适的刻度值，以便在图中展示数据。'''
fig, axes = plt.subplots(1, 4, figsize=(15,4),subplot_kw={'xticks':(), 'yticks':()})

# 列出要使用的算法
algorithms = [KMeans(n_clusters=2), AgglomerativeClustering(n_clusters=2), DBSCAN()]
# 创建一个随机的簇分配作为参考
# RandomState(seed=0) 创建了一个随机数生成器的实例，并将种子值（seed）设为 0。这个种子值决定了随机数生成器的初始状态。相同的种子值会产生相同的随机数序列。
random_state = np.random.RandomState(seed=0) 
'''在这里，low=0 表示最小可能的整数值是 0，high=2 表示最大可能的整数值是 1（不包含 2）。size=len(X) 意味着生成的随机整数数量与数据集 X 的长度相同。
所以，random_clusters 是一个长度与数据集 X 相同的一维数组，其中的每个元素都是随机选择的 0 或 1。'''
random_clusters = random_state.randint(low=0, high=2, size=len(X))

# 绘制随机分配
# X_scaled[:, 0]表示沿 x 轴的数据点的值;X_scaled[:, 1]表示沿 y 轴的数据点的值。c指定散点图中每个点的颜色,s散点标记大小
axes[0].scatter(X_scaled[:, 0], X_scaled[:, 1], c=random_clusters, cmap=mglearn.cm3, s=60)
# set_title是用于设置子图标题，adjusted_rand_score用于计算与聚类相关的某些评估指标的函数
axes[0].set_title("Random assignment - ARI: {:.2f}".format(adjusted_rand_score(y, random_clusters)))

for ax, algorithm in zip(axes[1:], algorithms):
    clusters = algorithm.fit_predict(X_scaled)
    # 绘制簇分配和簇中心
    # 该c=clusters参数根据上一步中获得的聚类标签为每个数据点分配颜色。
    ax.scatter(X_scaled[:, 0], X_scaled[:, 1], c=clusters, cmap=mglearn.cm3, s=60)
    # 算法的名称 ( algorithm.__class__.__name__) ，聚类的轮廓分数 ( silhouette_score(X_scaled, clusters))。
    ax.set_title("{} - ARI : {:.2f}".format(algorithm.__class__.__name__, adjusted_rand_score(y, clusters)))
```

调整rand 指数给出了符合直觉的结果，随机分配的分数为0，而DBSCAN(完美地找到了期望中的聚类)的分数为 1。

用这种方式评估聚类时，一个常见的错误是使用 accuracy_score 而不是adjusted_rand_score、normalized_mutual_info_score或其他聚类指标。使用精度的问题在于，它要求分配的簇标签与真实值完全匹配。但簇标签本身毫无意义一一唯一重要的是哪些点位于同一个簇中。

```python
from sklearn.metrics import accuracy_score

# 这两种点标签对应于相同的聚类
clusters1 = [0, 0, 1, 1, 0]
clusters2 = [1, 1, 0, 0, 1]
# 精度为0，因为二者标签完全不同
print("Accuracy: {:.2f}".format(accuracy_score(clusters1, clusters2)))
# 调整rand分数为1，因为二者聚类完全相同
print("ARI :{:.2f}".format(adjusted_rand_score(clusters1, clusters2)))
```

## 2. 在没有真实值的情况下评估聚类

我们刚刚展示了一种评估聚类算法的方法，但在实践中，使用诸如 ARI之类的指标有一个很大的问题。在应用聚类算法时，通常没有真实值来比较结果。如果我们知道了数据的正确聚类，那么可以使用这一信息构建一个监督模型 (比如分类器)。因此，使用类似ARI和NMI的指标通常仅有助于开发算法，但对评估应用是否成功没有帮助。

有一些聚类的评分指标不需要真实值，比如轮廓系数 (silhouette coeffcient)。但它们在实践中的效果并不好。轮廓分数计算一个簇的紧致度，其值越大越好，最高分数为 1。虽然紧致的簇很好，但紧致度不允许复杂的形状。

```python
# 利用无监督的轮廓分数在two_moons数据集上比较随机分配k均值、凝聚聚类和DBSCAN(更符合直觉的DBSCAN的轮廓分数低于k均值找到的分配)

X, y = make_moons(n_samples=200, noise=0.05, random_state=0)
# 将数据缩放为品均值为0，方差为1
scaler = StandardScaler()
scaler.fit(X)
X_scaled = scaler.transform(X)

fig, axes = plt.subplots(1, 4, figsize=(15,4), subplot_kw={'xticks':(), 'yticks':()})

# 创建一个随机的簇分配作为参考
random_state = np.random.RandomState(seed=0) # 创建一个随机数生成器的实例，并将种子值（seed）设为 0
random_clusters = random_state.randint(low=0, high=2, size=len(X)) # 使用特定随机状态生成随机整数的方法

# 绘制随机分配
axes[0].scatter(X_scaled[:, 0], X_scaled[:, 1], c=random_clusters, cmap=mglearn.cm3, s=60)
axes[0].set_title("Random assignment: {:.2f}".format(silhouette_score(X_scaled, random_clusters)))

# 列出使用算法
algorithms = [KMeans(n_clusters=2), AgglomerativeClustering(n_clusters=2),DBSCAN()]

for ax, algorithm in zip(axes[1:], algorithms):
    clusters = algorithm.fit_predict(X_scaled)
    # 绘制簇分配和簇中心
    # 该c=clusters参数根据上一步中获得的聚类标签为每个数据点分配颜色。
    ax.scatter(X_scaled[:, 0], X_scaled[:, 1], c=clusters, cmap=mglearn.cm3, s=60)
    # 算法的名称 ( algorithm.__class__.__name__) ，聚类的轮廓分数 ( silhouette_score(X_scaled, clusters))。
    ax.set_title("{}:{:.2f}".format(algorithm.__class__.__name__, silhouette_score(X_scaled, clusters)))
```

如你所见，k 均值的轮廓分数最高，尽管我们可能更喜欢 DBSCAN的结果。对于评估聚类，稍好的策略是使用基于鲁棒性的 (robustness-based)聚类指标。这种指标先向数据中添加一些噪声，或者使用不同的参数设定，然后运行算法，并对结果进行比较。其思想是，如果许多算法参数和许多数据扰动返回相同的结果，那么它很可能是可信的。

即使我们得到一个鲁棒性很好的聚类或者非常高的轮廓分数，但仍然不知道聚类中是否有任何语义含义，或者聚类是否反映了数据中我们感兴趣的某个方面。我们回到人脸图像的例子。我们希望找到类似人脸的分组，比如男人和女人、老人和年轻人，或者有胡子的人和没胡子的人。假设我们将数据分为两个簇，关于哪些点应该被聚类在一起，所有算法的结果一致。我们仍不知道找到的簇是否以某种方式对应于我们感兴趣的概念。算法找到的可能是侧视图和正面视图、夜间拍摄的照片和白天拍摄的照片，或者 iPhone 拍摄的照片和安卓手机拍摄的照片。要想知道聚类是否对应千我们感兴趣的内容，唯一的办法就是对管进行人工分析。

## 3.在人脸数据集上比较算法

将k均值、DBSCAN和凝聚聚类算法应用于 Wid 数据集中的 Labeled Faces，并查看它们是否找到了有趣的结构。使用数据的特征脸表示，它由包含 100 个成分的PCA(whiten=True)生成:

```python
from sklearn.decomposition import PCA
from sklearn.datasets import fetch_lfw_people
from sklearn.model_selection import train_test_split

# 从lfw数据集中提取特征脸，并对数据进行变换

'''
// people = fetch_lfw_people(data_home='E:/clong/machine-learning/data/', min_faces_per_person=20, resize=0.7)
fetch_lfw_people是 scikit-learndatasets模块中的一个函数，用于加载 LFW 数据集。
data_home指定存储数据集的目录。
min_faces_per_person=20设置一个条件，仅包含数据集中至少有 20 张图像的人。
resize=0.7将图像大小调整为原始大小的 70%。


// image_shape = people.images[0].shape
该行提取数据集中第一张图像的形状。它将提供有关图像尺寸的信息（例如，高度、宽度和颜色通道数）。


// mask = np.zeros(people.target.shape, dtype=np.bool_)
此行创建与 形状相同的布尔掩码people.target，并使用所有False值进行初始化。该掩码将用于从数据集中选择数据点的子集。


// for target in np.unique(people.target):
此循环迭代数据集中的唯一目标标签。每个目标标签对应于数据集中的不同人。


// mask[np.where(people.target == target)[0][:50]] = 1
对于每个唯一的目标标签，此行使用 来标识属于该人的数据点（图像）的索引np.where(people.target == target)[0]。
[0][:50]选择该人的前 50 个索引。这用于将每人的图像数量限制为 50 张。
最后，mask更新True为那些选定的索引。

// X_people = people.data[mask]
X_people此行通过将布尔掩码应用于mask数据来创建数据的子集。这将根据循环中设置的条件选择图像子集。


// y_people = people.target[mask]
y_people该行基于相同的布尔掩码创建目标标签的相应子集mask。这确保标签对应于选定的图像子集。

总之，此代码加载 LFW 数据集，调整图像大小，然后创建数据子集，每人最多包含 50 张图像，每人至少包含 20 张图像。
该子集存储在 中X_people，相应的目标标签存储在 中y_people。布尔掩码mask用于选择适当的数据点子集。
'''
people = fetch_lfw_people(data_home='E:/clong/machine-learning/data/',min_faces_per_person=20, resize=0.7)
image_shape = people.images[0].shape

mask = np.zeros(people.target.shape,dtype=np.bool_)
for target in np.unique(people.target):
    mask[np.where(people.target == target)[0][:50]] = 1

X_people = people.data[mask]
y_people = people.target[mask]

# 将灰度值缩放到0和1之间，而不是0到255之间
# 以得到更好的数据
X_people= X_people / 255

# 将数据分为训练集和测试集
X_train, X_test, y_train, y_test = train_test_split(X_people, y_people, 
                                                    stratify=y_people,random_state=0)

# 使用指定的配置初始化 PCA 对象，包括保留 100 个主成分、白化转换后的数据以及设置随机种子以实现再现性。
pca = PCA(n_components=100, whiten=True, random_state=0)
pca.fit_transform(X_people)
X_pca = pca.transform(X_people)
# 代码在同一数据集上执行两次 PCA 降维。第一次（在 中pca.fit_transform(X_people)），计算变换但不存储。
# 第二次（在 中X_pca = pca.transform(X_people)），应用变换并将其存储在 中X_pca。
```

我们之前见到，与原始像素相比，这是对人脸图像的一种语义更强的表示。它的计算速度也更快。这里有一个很好的练习，就是在原始数据上运行下列实验、不要用 PCA，并观察你是否能找到类似的簇。

### 用DBSCAN分析人脸数据集:

```python
dbscan = DBSCAN()
labels = dbscan.fit_predict(X_pca)
print("Unique labels: {}".format(np.unique(labels)))
```

我们看到，所有返回的标签都是 -1，因此所有数据都被 DBSCAN标记为“噪声”。可以改变两个参数来改进这一点:第一，我们可以增大 eps，从而扩展每个点的邻域；第二，我们可以减小mn_samples，从而将更小的点组视为簇。首先尝试改变min samples:

```python
dbscan = DBSCAN(min_samples=3)
labels = dbscan.fit_predict(X_pca)
print("Unique labels: {}".format(np.unique(labels)))
```

使仅考虑由三个点构成的组，所有点也都被标记为噪声。因此需要增大 eps:

```python
dbscan = DBSCAN(min_samples=3, eps=15)
labels = dbscan.fit_predict(X_pca)
print("Unique labels: {}".format(np.unique(labels)))
```

使用更大的 eps(其值为15)，我们只得到了单一簇和噪声点。我们可以利用这一结果找出“噪声”相对于其他数据的形状。为了进一步理解发生的事情，我们查看有多少点是噪声有多少点在簇内:

```python
# 计算所有簇中的点数和噪声中的点数
# bincount 不允许负值，所以我们需要加1
# 结果中的第一个数字对应于噪声点
print("Number of points per cluster:{}".format(np.bincount(labels + 1)))
```

噪声点非常少一-只有 27个，因此我们可以查看所有的噪声点`

```python
%matplotlib inline
noise = X_people[labels==-1]
fig, axes = plt.subplots(3, 9, subplot_kw={'xticks':(), 'yticks':()},figsize=(12,4))
for image, ax in zip(noise, axes.ravel()):
    ax.imshow(image.reshape(image_shape), vmin=0, vmax=1)
```

将这些图像与前面中随机选择的人脸图像样本进行比较，我们可以猜测它们被标记为噪声的原因:第1行第5张图像显示一个人正在用玻璃杯喝水，还有人戴帽子的图像.在最后一张图像中，人脸前面有一只手。其他图像都包含奇怪的角度，或者太近或太宽的剪切。

这种类型的分析一-尝试找出“奇怪的那一个”一被称为**异常值检测**(outlierdetection)。如果这是一个真实的应用，那么我们可能会尝试更好地裁切图像，以得到更加均匀的数据。对于照片中的人有时戴着帽子、喝水或在面前举着某物，我们能做的事情很少。但需要知道它们是数据中存在的问题，我们应用任何算法都需要解决这些问题。

如果我们想要找到更有趣的簇，而不是一个非常大的簇，那么需要将 eps 设置得更小，取值在15和0.5(默认值)之间。我们来看一下eps 不同取值对应的结果:

```python
for eps in [1, 3, 5, 7, 9, 11, 13]:
    print("\neps = {}".format(eps))
    dbscan = DBSCAN(eps=eps, min_samples=3)
    labels = dbscan.fit_predict(X_pca)
    print("Clusters present: {}".format(np.unique(labels)))
    print("Clusters size: {}".format(np.bincount(labels + 1)))
```

对于较小的eps，所有点都被标记为噪声。eps=7时，我们得到许多噪声点和许多较小的簇。eps=9时，我们仍得到许多噪声点，但我们得到了一个较大的簇和一些较小的簇。从eps=11开始，我们仅得到一个较大的簇和噪声。

有趣的是，较大的簇从来没有超过一个。最多有一个较大的簇包含大多数点，还有一些较小的簇。这表示数据中没有两类或三类非常不同的人脸图像，而是所有图像或多或少地都与其他图像具有相同的相似度(或不相似度)。

eps=7的结果看起来最有趣，它有许多较小的簇。我们可以通过将13 个较小的簇中的点全部可视化来深人研究这一聚类

```python
dbscan = DBSCAN(eps=7, min_samples=3)
labels = dbscan.fit_predict(X_pca)

for cluster in range(max(labels) + 1):
    mask = labels == cluster
    n_image = np.sum(mask)
    fig, axes = plt.subplots(1, n_image, figsize=(n_image * 1.5, 4),subplot_kw={'xticks':(), 'yticks':()})
    for image, label, ax in zip(X_people[mask], y_people[mask], axes):
        ax.imshow(image.reshape(image_shape), vmin=0, vmax=1)
        ax.set_title(people.target_names[label].split()[-1])
```

在每个簇内，人脸方向和面部表情也是固定的。有些簇中包含多个人的面孔，但他们的方向和表情都相似。

这就是我们将 DBSCAN 算法应用于人脸数据集的分析结论。如你所见，我们这里进行了人工分析，不同于监督学习中基于R’分数或精度的更为自动化的搜索方法。

### 用k均值分析人脸数据集

利用DBSCAN 无法创建多于一个较大的簇。凝聚聚类和k均值更可能创建均匀大小的簇，但我们需要设置簇的目标个数。我们可以将簇的数量设置为数据集中的已知人数，虽然无监督聚类算法不太可能完全找到它们。相反，我们可以首先设置一个比较小的簇的数量，比如 10个，这样我们可以分析每个族:

```python
# 用k均值提取簇
km = KMeans(n_clusters=10, random_state=0)
labels_km = km.fit_predict(X_pca)
print("Clusters size k-means: {}".format(np.bincount(labels_km)))
```

如你所见，k均值聚类将数据划分为大小相似的簇，其大小在 70 和 424之间。这与DBSCAN的结果非常不同。

可以通过将簇中心可视化来进一步分析k 均值的结果。由于我们是在 PCA生成的表示中进行聚类，因此我们需要使用 pca.inverse transform 将中心旋转回到原始空间并可视化:

```python
fig, axes = plt.subplots(2, 5, subplot_kw={'xticks': (), 'yticks': ()},
                         figsize=(12, 4))
for center, ax in zip(km.cluster_centers_, axes.ravel()):
    ax.imshow(pca.inverse_transform(center).reshape(image_shape),
              vmin=0, vmax=1)
```

k均值找到的簇中心是非常平滑的人脸。这并不奇怪，因为每个簇中心都是 64到 386 张人脸图像的平均。使用降维的 PCA 表示，可以增加图像的平滑度。聚类似乎捕捉到人脸的不同方向、不同表情(第3 个簇中心似乎显示的是一张笑脸)，以及是否有衬衫领子(见倒数第二个簇中心)。

对每个簇中心给出了簇中5张最典型的图像 (该簇中与簇中心距离最近的图像)与5张最不典型的图像 (该族中与族中心距离最远的图像):

```python
mglearn.plots.plot_kmeans_faces(km, pca, X_pca, X_people,
                                y_people, people.target_names)
```

不过非典型的点与簇中心不太相似，而且它们的分配似乎有些随意。这可以归因于以下事实:k均值对所有数据点进行划分，不像 DBSCAN那样具有“噪声”点的概念。利用更多数量的簇，算法可以找到更细微的区别。但添加更多的簇会使得人工检查更加困难。

可以通过计算ARI来度量凝聚聚类和k均值给出的两种数据划分是否相似:

```python
print("ARI: {:.2f}".format(adjusted_rand_score(labels_agg, labels_km)))
```

ARI只有0.13，说明labels_agg和labels_k 这两种聚类的共同点很少。这并不奇怪，原
因在于以下事实:对于k均值，远离簇中心的点似乎没有什么共同点。

下面，我们绘制树状图。我们将限制图中树的深度，因为如果分支到 2063 个数据点，图像将密密麻麻无法阅读:

```python
# 聚聚类在人脸数据集上的树状图
from scipy.cluster.hierarchy import dendrogram, ward
linkage_array = ward(X_pca)
# 现在我们为包含簇之间距离的link_array绘制树状图
plt.figure(figsize=(20,5))
dendrogram(linkage_array, p=7, truncate_mode='level', no_labels=True)
plt.xlabel("Sample index")
plt.ylabel("Cluster distance")
```

    要想创建 10个簇，我们在顶部有 10 条竖线的位置将树横切。在上面图所示的玩具数据的树状图中，你可以从分支的长度中看出，两个或三个簇就可以很好地划分数据。对于人脸数据而言，似乎没有非常自然的切割点。有一些分支代表更为不同的组，但似乎没有一个特别合适的簇的数量。这并不奇怪，因为 DBSCAN的结果是试图将所有的点都聚类在一起。
