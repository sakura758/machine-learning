# 预处理与缩放

一些算法(如神经网络和SVM) 对数据缩放非常敏感。因此，通常的做法是对特征进行调节，使数据表示更适合于这些算法。通常来说，这是对数据的一种简单的按特征的缩放和移动。

```python
import mglearn

mglearn.plots.plot_scaling()
# 对数据集缩放和预处理的各种方法
```

## 1. 不同类型的预处理

scikit-learn中的StandardScaler 确保每个特征的平均值为0、方差为 1，使所有特征都位于同一量级。但这种缩放不能保证特征任何特定的最大值和最小值。RobustScaler 的工作原理与standardScaler 类似，确保每个特的统计属性都位于同一范围。但 RobustScaler 使用的是中位数和四分位数”而不是平均值和方差。这样 RobustScaler 会忽略与其他点有很大不司的数据点 (比如测量误差)。这些与众不同的数据点也叫异常值 (outlier)，可能会给其他缩放方法造成麻烦。
  

见：scikit-learn数据预处理之标准化.md

## 2. 应用数据变换


通常在应用监督学习算法之前使用预处理方法 (比如缩放)

```python
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler

cancer = load_breast_cancer()
X_train, X_test, y_train, y_test = train_test_split(cancer.data, cancer.target, random_state=1)
scaler = MinMaxScaler()
# 使用fit 方法拟合缩放器 (scaler)，并将其应用于训练数据。
# 对于MnMaxScaler 来说，fit 方法计算训练集中每个特征的最大值和最小值。
# 与前面的分类器和回归器(regressor)不同，在对缩放器调用fit时只提供了x_train，而不用ytrain:
scaler.fit(X_train)
# 变换数据
X_train_scaler = scaler.transform(X_train)
# 在缩放之前和之后分别打印数据集的属性
print("transformed shape: {}\n".format(X_train_scaler.shape))
print('-*_'*30)
print("per-feature minimum before scaling:\n{}\n".format(X_train.min(axis=0)))
print('-*_'*30)
print("per-feature maximum before scaling:\n{}\n".format(X_train.max(axis=0)))
print('-*_'*30)
print("per-feature minimum after scaling:\n{}\n".format(X_train_scaler.min(axis=0)))
print('-*_'*30)
print("per-feature maximum after scaling :\n{}".format(X_train_scaler.max(axis=0)))
```

```python
# 对测试数据进行变换
X_test_scaler = scaler.transform(X_test)
# 在缩放之后打印测试数据的属性
print("per-feature minimum after scaling:\n{}".format(X_test_scaler.min(axis=0)))
print('\n')
print("per-feature maximum after scaling:\n{}".format(X_test_scaler.max(axis=0)))
```

对测试集缩放后的最大值和最小值不是 1和0，这或许有些出乎意料。有些特征甚至在0~1的范围之外!对此的解释是，MinMaxScaler (以及其他所有缩放器)总是对训练集和测试集应用完全相同的变换。也就是说，transform 方法总是减去训练集的最小值，然后除以训练集的范围，而这两个值可能与测试集的最小值和范围并不相同。

## 3.对训练数据和测试数据进行相同的缩放

```python
from sklearn.datasets import make_blobs
from matplotlib import pyplot as plt
from sklearn.preprocessing import MinMaxScaler

# 构造数据
# cluster_std是make_blobs函数中的一个参数，它控制了生成的聚类数据集中每个簇中数据点的分布范围或者说波动程度。
# 具体来说，cluster_std代表了每个簇中数据点与其所在簇的中心点之间的标准差。
# 标准差越大，数据点分布越分散，簇的形状越模糊；标准差越小，数据点分布越集中，簇的形状越清晰。

X, _ = make_blobs(n_samples=50, centers=5, random_state=4, cluster_std=2)
# 将其分为训练集和测试集
# test_size=.1: 这指定了测试集的大小。在这里，.1 表示将数据集的 10% 分配给测试集，剩下的 90% 用于训练集。
X_train, X_test= train_test_split(X, random_state=5, test_size=.1)

# 绘制训练集和测试集
fig, axes = plt.subplots(1, 3, figsize=(16,5))
# c 参数用于指定散点的颜色,mglearn.cm2(0) 似乎是一个用于生成颜色映射的函数，通过参数0来选择颜色。
# s=60: 这是散点的大小，指定为60个像素。
axes[0].scatter(X_train[:, 0], X_train[:, 1], c=mglearn.cm2(0), label="TRaining set", s=60)
axes[0].scatter(X_test[:, 0], X_test[:, 1], marker='^', c=mglearn.cm2(1), label="Test set", s=60)
axes[0].legend(loc="upper left")  # 添加图例
axes[0].set_title('Original Data')

# 利用MinMaxScaler缩放数据
scaler = MinMaxScaler()
scaler.fit(X_train)
X_train_scaler = scaler.transform(X_train)
X_test_scaler = scaler.transform(X_test)

# 将正确缩放的数据可视化
axes[1].scatter(X_train_scaler[:, 0], X_train_scaler[:, 1], c=mglearn.cm2(0), label="TRaining set", s=60)
axes[1].scatter(X_test_scaler[:, 0], X_test_scaler[:, 1], marker='^', c=mglearn.cm2(1), label="Test set", s=60)
axes[1].legend(loc="upper left")  # 添加图例
axes[1].set_title('Scaled Data')



# 错误缩放
#单独对测试集进行缩放
# 错误
test_scaler = MinMaxScaler()
test_scaler.fit(X_test)
X_false = test_scaler.transform(X_test)

# 错误可视化
axes[2].scatter(X_train_scaler[:, 0], X_train_scaler[:, 1], c=mglearn.cm2(0), label="TRaining set", s=60)
axes[2].scatter(X_false[:, 0], X_false[:, 1], marker='^', c=mglearn.cm2(1), label="Test set", s=60)
axes[2].legend(loc="upper left")  # 添加图例
axes[2].set_title('False Data')

for ax in axes:
    ax.set_xlabel("Feature 0")
    ax.set_ylabel("Feature 1")
fig.tight_layout()
```

第三张图展示了如果我们对训练集和测试集分别进行缩放会发生什么。在这种情况下，对训练集和测试集而言，特征的最大值和最小值都是1和0。但现在数据集看起来不一样。测试集相对训练集的移动不一致，因为它们分别做了不同的缩放。我们随意改变了数据的排列

## 4. 预处理对监督学习的作用

```python
# 在原始数据上拟合svc
from sklearn.svm import SVC

X_train, X_test, y_train, y_test = train_test_split(cancer.data, cancer.target, random_state=0)
svm= SVC()
svm.fit(X_train, y_train)
print("Test set accuary: {:.3f}".format(svm.score(X_test, y_test)))
```

```python
# MinMaxScaler 对数据进行缩放，然后再拟合SVC:
scaler = MinMaxScaler()
scaler.fit(X_train)
X_train_scaler = scaler.transform(X_train)
X_test_scaler = scaler.transform(X_test)

# 在缩放后的数据上学习SVM
svm.fit(X_train_scaler, y_train)

# 计算分数
print("test accuary: {:.3f}".format(svm.score(X_test_scaler, y_test)))
```

也可以通过改变使用的类将一种预处理算法轻松替换成另一种，因为所有的预处理类都具有相同的接口，都包含fit和transform方法
