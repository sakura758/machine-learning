```python
X,y = mglearn.datasets.make_forge()
X_train,X_test,y_train,y_test = train_test_split(X,y,random_state=0)
# 当将数据集划分为训练集和测试集时，通常使用 random_state 来确保每次划分都是相同的。这对于比较不同模型的性能或进行交叉验证很重要。
```

```python
print(clf.predict(X_test))
# 为了评估模型的泛化能力好坏，我们可以对测试数据和测试标签调用 score 方法
print(clf.score(X_test,y_test))
```

### （2）分析KNeighborsClassifier

```python
# 分别将1个、3个和9个邻居决策边 边界可视化
fig ,axes = plt.subplots(1,3,figsize = (10,3))  # 创建一个包含3个子图的图形窗口（fig），这三个子图将在一行中排列
for n_neighbors, ax in zip([1,3,9],axes):
# fit方法返回实例对象，可以将实例化和拟合放在一起
    clf = KNeighborsClassifier(n_neighbors=n_neighbors).fit(X,y)
    mglearn.plots.plot_2d_classification(clf, X, fill=True, eps=0.5, ax=ax, alpha=.4)
    mglearn.discrete_scatter(X[:,0], X[:,1], y, ax=ax)
    ax.set_title("{} neighbor(s)".format(n_neighbors))
    ax.set_xlabel("feature 0")  # Corrected from set_xlable to set_xlabel
    ax.set_ylabel("feature 1")  # Corrected from set_ylable to set_ylabel
axes[0].legend(loc=3)
plt.show()
```

使用更少的邻居对应更高的模型复杂度，而使用更多的邻居对应更低的模型复杂度。

将在现实世界的乳腺癌数据集上进行研究。先将数据集分成训练集和测试集，然后用不同的邻居个数对训练集和测试集的性能进行评估。

```python
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
```

```python
cancer = load_breast_cancer()  #加载数据
print(cancer.keys())
print(cancer.data[:2])
print(cancer.target[:10])
print(cancer.data.shape)
print(cancer.target.shape)
```

```python
X_train, X_test, y_train, y_test = train_test_split(cancer.data, cancer.target, stratify=cancer.target, random_state=66)
# stratify 来保证训练集和测试集中的类别分布是相似的
# random_state 是一个用于指定随机数生成器种子的参数。
# X: 特征矩阵或数据集（通常是一个二维数组或 DataFrame）。
# y: 目标变量或标签。
# ##test_size: 测试集的大小，可以是一个浮点数（表示比例）或整数（表示样本数量）。
```

```python
train_accuracy = []
test_accuracy = []
neighbors_setting = range(1,11)

for n_neighbors in neighbors_setting:
    # 构建模型
    clf = KNeighborsClassifier(n_neighbors=n_neighbors)
    clf.fit(X_train, y_train)
    # 训练精度
    train_accuracy.append(clf.score(X_train, y_train))
    # 泛化精度
    test_accuracy.append(clf.score(X_test, y_test))

# 画图
plt.plot(neighbors_setting, train_accuracy, label="train accuracy")  # Corrected from 'lable' to 'label'
plt.plot(neighbors_setting, test_accuracy, label="test accuracy")  # Corrected from 'lable' to 'label'
plt.xlabel("n_neighbors")
plt.ylabel("accuracy")
plt.legend()
plt.show()
```

单邻居时的测试集精度比使用更多邻居时要低，这表示单一近邻的模型过于复杂。与之相反，当考虑 10个邻居时，模型又过于简单，性能甚至变得更差。    
最佳性能在中间的某处邻居个数大约为6。
